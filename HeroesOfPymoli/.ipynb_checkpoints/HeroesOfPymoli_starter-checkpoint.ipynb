```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchase_df = pd.DataFrame(purchase_data)
player_count = purchase_df["SN"].nunique()
print("Total Number of Players: " + str(player_count))
purchase_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_df["Item Name"].nunique()


avg_price = round(purchase_df["Price"].mean(),2)


total_purchases = purchase_df["Purchase ID"].count()


total_rev = purchase_df["Price"].sum()


my_summary = [{"Unique Items": int(unique_items),
              "Average Price": (avg_price),
               "Total Purchases": int(total_purchases),
              "Total Revenue": (total_rev)
              }]

summary_df = pd.DataFrame(my_summary)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender Demographics

gender_df = purchase_df[["Gender", "SN"]].drop_duplicates(subset=["SN"], keep="first")
gender_df
gender_count = gender_df["Gender"].value_counts()
gender_pct = round(gender_df["Gender"].value_counts(normalize = True),2)
pd.concat([gender_count,gender_pct], axis=1, keys=['Total Count',"Percentage"])
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Basic Calculations for Gender

gender_purchasing_df = purchase_df[["Gender", "SN", "Price"]].groupby(["Gender"])

purchase_count = gender_purchasing_df["Price"].count()

purchase_avg = gender_purchasing_df["Price"].mean()

total_purchase = gender_purchasing_df["Price"].sum()

person_purchase = gender_purchasing_df.mean()

gender_purchase_analysis = pd.concat([purchase_count,purchase_avg, total_purchase],
          axis=1, keys=["Total Purchases", "Avg. Purchase Price", "Purchase Total"])

gender_purchase_analysis
```

```python
#Basic Calculations for Gender & Individual Person

person_purchasing_df = purchase_df[["Gender", "SN", "Price"]].groupby(["Gender", "SN"])

purchase_count = person_purchasing_df["Price"].count()

purchase_avg = round(person_purchasing_df["Price"].mean(),2)

total_purchase = person_purchasing_df["Price"].sum()


gender_purchase_analysis = pd.concat([purchase_count,purchase_avg, total_purchase],
          axis=1, keys=["Total Purchases", "Avg. Purchase Price", "Purchase Total"])

gender_purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [5,9,13,17,21,25,29,33,37,41,45]

#Creates column in DF for Age Bin assignments
purchase_df["Age Bins"] = pd.cut(purchase_df["Age"], age_bins)
purchase_df

#only allows unique players information and calculates count/percentage
age_demo_df = purchase_df[["Age Bins", "SN"]].drop_duplicates(subset=["SN"], keep="first")
age_count = age_demo_df["Age Bins"].value_counts()
age_pct = round(age_demo_df["Age Bins"].value_counts(normalize = True),2)

#builds dataframe to display
age_demographics = pd.concat([age_count, age_pct], axis=1, keys=["Total Count", "Percentage"])
age_demographics


#gender_count = gender_df["Gender"].value_counts()
#gender_pct = gender_df["Gender"].value_counts(normalize = True)
#pd.concat([gender_count,gender_pct], axis=1, keys=['Total Count',"Percentage"])
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [5,9,13,17,21,25,29,33,37,41,45]

purchase_df["Age Bins"] = pd.cut(purchase_df["Age"], age_bins)
purchase_df
grouped_age_df = purchase_df.groupby(["Age Bins"])

purchase_count = grouped_age_df["Price"].count()

purchase_avg = round(grouped_age_df["Price"].mean(),2)

total_purchase = grouped_age_df["Price"].sum()

person_purchase = grouped_age_df.mean()

age_purchase_analysis = pd.concat([purchase_count,purchase_avg, total_purchase],
          axis=1, keys=["Total Purchases", "Avg. Purchase Price", "Total Purchase Value"])

age_purchase_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_count = purchase_df[["SN","Price"]].groupby(["SN"]).count().sort_values("Price", ascending = False)
top_spenders_count.head(5)

top_spenders_total = purchase_df[["SN","Price"]].groupby(["SN"]).sum().sort_values("Price", ascending = False)
top_spenders_total

top_spenders_avg = purchase_df[["SN","Price"]].groupby(["SN"]).mean().sort_values("Price", ascending = False)

top_spender_analysis = pd.concat([top_spenders_count, top_spenders_total, round(top_spenders_avg,2)],
          axis=1, keys=["Purchase Count", "Total Purchase Value", "Average Purchase Price"])

top_spender_analysis

```
